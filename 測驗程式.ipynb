```python
import json
import os

# 定義儲存資料的 JSON 檔案路徑
file_path = "data.json"

# 如果檔案已經存在，先讀取現有資料
if os.path.exists(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        data = json.load(f)
else:
    data = []

def get_input():
    entry = {}

    # 輸入日期（四位數字）
    date = input("請輸入日期(四位數字，例如1010): ")
    entry["date"] = date

    # 輸入句子
    sentence = input("請輸入句子: ")
    entry["sentence"] = sentence

    # 輸入第一個意思
    meaning1 = input("請輸入第一個意思: ")
    entry["meaning1"] = meaning1

    # 輸入四聲拼音
    pinyin1 = input("請輸入句子的四聲拼音: ")
    entry["pinyin1"] = pinyin1

    # 輸入海報拼音
    pinyin_poster1 = input("請輸入句子的海報拼音: ")
    entry["pinyin_poster1"] = pinyin_poster1

    return entry

while True:
    entry = get_input()

    # 檢查日期是否已經存在
    existing_entry = next((item for item in data if item["date"] == entry["date"]), None)
    if existing_entry:
        print(f"已存在的句子: {existing_entry['sentence']}")
        update = input("是否要修改[y/n]: ").lower()
        if update == 'y':
            print("正在修改已有的資料...")
            existing_entry.update(entry)
        else:
            print("跳過修改")
    else:
        data.append(entry)

    # 輸入下一個資料
    next_entry = input("是否要繼續輸入下一個資料？(y/n): ").lower()
    if next_entry != 'y':
        break

# 將資料儲存到 data.json 檔案
with open(file_path, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

print(f"資料已儲存到 {file_path} 檔案中")
```
